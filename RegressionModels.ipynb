```python
# Import librarires

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generate synthetic data 
X = 5 * np.random.rand(100,1) 
Y=  5 *X  + 4 + np.random.randn(100,1) 
plt.scatter (X,Y)
plt.xlabel('X') 
plt.ylabel('Y')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()
```

```python
#from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,Y)
X_vals = 5 * np.linspace (0,1,100).reshape(-1,1)
Y_vals = reg.predict(X_vals)

plt.scatter (X,Y)
plt.plot(X_vals, Y_vals, color = "r" )
plt.xlabel('X') 
plt.ylabel('Y')
plt.ylim(ymin=0)
plt.xlim(xmin=0)
plt.show()
```

```python
#display regression coefficients and R-squared value of model
print("Intercept: " , reg.intercept_)
print("Slope: "     , reg.coef_)
print("R^2: "       , reg.score(X, Y))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(california.data, california.target, test_size = 0.20, random_state=45)
print(X_train.shape)
print (X_test.shape)
print (y_train.shape)
```

```python
# from sklearn.linear_model import LinearRegression
linear_regression = LinearRegression()
linear_regression.fit(X=X_train, y=y_train)
```

```python
for i, name in enumerate(california.feature_names):
    print(f'{name:>10}: {linear_regression.coef_[i]}')

print (linear_regression.intercept_)
print(linear_regression.score(X_train, y_train))
```

```python
# from sklearn.metrics import mean_squared_error as mse
from sklearn import metrics

predicted = linear_regression.predict(X_test)
expected = y_test
df = pd.DataFrame()
df['Expected'] = pd.Series(expected)
df['Predicted'] = pd.Series(predicted)

print("R^2: ", metrics.r2_score(expected, predicted))
print("MSE: ", metrics.mean_squared_error(expected, predicted))

figure = plt.figure(figsize=(9, 9))
axes = sns.scatterplot(data=df, x='Expected', y='Predicted', \
                       hue='Predicted', palette='cool', legend=False)
start = min(expected.min(), predicted.min())
end = max(expected.max(), predicted.max())
axes.set_xlim(start, end)
axes.set_ylim(start, end)
line = plt.plot([start, end], [start, end], 'k--')
```

```python
dataset_iris = datasets.load_iris()
#dataset_iris
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
target = 'species'
iris = pd.DataFrame(
    dataset_iris.data,
    columns=features)
iris[target] = dataset_iris.target
iris.head()
```

```python
iris.sample (10)
```

```python
from matplotlib.colors import ListedColormap
sns.set()
sns.set(rc={"figure.figsize": (10, 8)})
PALETTE = sns.color_palette('deep', n_colors=3)
CMAP = ListedColormap(PALETTE.as_hex())
RANDOM_STATE = 42
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

pca = PCA(n_components=2)
points_pca = pca.fit_transform(iris[features])

tsne = TSNE(n_components=2, n_iter=1000, random_state=RANDOM_STATE)
points_tsne = tsne.fit_transform(iris[features])
#points_tsne
```

```python
plot_iris_2d(
    x = points_pca[:,0],
    y = points_pca[:,1],
    title = 'Iris dataset visualized with PCA')
```

```python
plot_iris_2d(
    x = points_tsne[:, 0],
    y = points_tsne[:, 1],
    title = 'Iris dataset visualized with t-SNE')
```

```python
def plot_iris_2d(x, y, title, xlabel="1st eigenvector", ylabel="2nd eigenvector"):
    sns.set_style("darkgrid")
    
    plt.scatter(x, y,
        c=iris['species'],
        cmap=CMAP,
        s=70)
    
    plt.title(title, fontsize=20, y=1.03)

    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    #plt.legend(handles=scatter.legend_elements()[0], 
    #       title="species")
```
